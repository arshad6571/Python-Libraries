```python
batters = pd.read_csv('/content/batter.csv')
batters.head(5)
```

```python
sample =batters.head(100).sample(25,random_state=5)
sample.head(5)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(sample['avg'],sample['strike_rate'])
```

```python

#now we dont know each dot represent which player for this we will execute a loop
plt.figure(figsize=(15,15))
plt.scatter(sample['avg'],sample['strike_rate'],s=sample['runs'])
for i in range(sample.shape[0]):
  plt.text(sample['avg'].values[i],sample['strike_rate'].values[i],sample['batter'].values[i])
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')


ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
```

```python
import matplotlib.pyplot as plt

#ploting more graph in one frame

fig, ax = plt.subplots(nrows=2,ncols=2,sharex=False,figsize=(10,6)) # Changed ncols to 2 and sharex to False

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'])

ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
```

```python
#another way to plot multiple figs in one frame

fig = plt.figure()

ax1 = fig.add_subplot(2,1,1) #targeting first graph having 2 rows and 1 cols (2,1,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')
ax1.set_title('Avg Vs Strike Rate')
ax1.set_ylabel('Strike Rate')
ax2 = fig.add_subplot(2,1,2) # targeting second graph order 212, means two rows one col and fig 2
ax2.scatter(batters['avg'],batters['runs'])

ax2.set_ylabel('Runs')
```

```python
#3D ploting
plt.figure(figsize=(15,15))


fig = plt.figure()
ax = plt.subplot(projection='3d')
ax.scatter3D(batters['avg'],batters['strike_rate'],batters['runs'])
ax.set_title('Avg Vs Strike Rate Vs Runs')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')
ax.set_zlabel('Runs')
```

```python
x = np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
xx,yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
```

```python
fig = plt.figure()
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z)
plt.colorbar(p)
```

```python
#contour plot: which convert a 3D plot into 2D
fig = plt.figure(figsize=(12,6))
fig = plt.figure()
ax = plt.subplot()
p = ax.contour(xx,yy,z)
plt.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,6))
fig = plt.figure()
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap= 'viridis')
plt.colorbar(p)
```
