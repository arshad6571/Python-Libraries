```python
#union1d----> just like maths make a new array of ele without duplication
e = np.array((1,2,3,4,5,6,7,5,4,3,2))
f = np.array((11,12,13,4,5,16,7,5,41,3,12))
np.union1d(e,f)
```

```python
# intersect1d ----> intersect means take only common
e = np.array((1,2,3,4,5,6,7,5,4,3,2))
f = np.array((11,12,13,4,5,16,7,5,41,3,12))
np.intersect1d(e,f)
```

```python
# setdiff1d -----> set didderence means select only those value which are
# avaible in set e but not in f
np.setdiff1d(e,f)
```
