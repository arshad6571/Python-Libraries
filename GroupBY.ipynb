```python
#Applying aggregate function{min,max,mean,std} in once
#passing dictionary
genre.agg(
    {
        'Runtime':'min',
        'Gross':'max',
        'IMDB_Rating':'mean',
        'No_of_Votes':'std'
    }
)
```

```python
#passing list

# Select only numeric columns from the DataFrame *before* applying the agg functions
numeric_columns_to_agg = movies.select_dtypes(include=np.number).columns

genre[numeric_columns_to_agg].agg(
    [
        'min',
        'max',
        'mean',
        'sum'
    ]
)
```

```python
#loops in groups
max_rating_movies = []
for groups, data in genre:
    max_rating_movies.append(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])
df = pd.concat(max_rating_movies, ignore_index=True)
df
```

```python
#Apply function
genre.apply('min')
```

```python
# find number of movies starting with A for each group
def foo(group):
  return group['Series_Title'].str.startswith('A').sum()
```

```python
genre.apply(foo)
```

```python
# find ranking of each movie in the group according to IMDB score
def my_rank(group):
  group['rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genre.apply(my_rank)
```

```python
# find normalized IMDB rating group wise
def rating(groups):
  groups['norm_rat'] = (groups['IMDB_Rating'].min() / groups['IMDB_Rating'].max())
  return groups
```

```python
genre.apply(rating)
```

```python
ipl = pd.read_csv('/content/deliveries.csv')
ipl.head(3)
```

```python
ipl.shape
```

```python
# find the top 10 batsman in terms of runs

ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no of sixes
sixes = ipl[ipl['batsman_runs'] == 6]
sixes.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head()
```

```python
# find V Kohli's record against all teams
temp_df = ipl[ipl['batsman'] == 'V Kohli']

temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# Create a function that can return the highest score of any batsman

def highest(batsman):
  temp_df = ipl[ipl['batsman'] == batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
highest('V Kohli')
```
