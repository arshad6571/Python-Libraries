#pie chart
### Pie Chart

image.png

Univariate/Bivariate Analysis

Categorical vs numerical

Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
df7 = pd.read_csv('/content/gayle-175.csv')
df7.head(5)
```

```python
plt.pie(df7['batsman_runs'],labels=df7['batsman'])
plt.show()
```

```python
# percentage and colors
plt.pie(df7['batsman_runs'],labels=df7['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df7['batsman_runs'],labels=df7['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True) #only first and last  value out from chart
plt.show()
```
