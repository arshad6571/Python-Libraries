**Pandas** is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language.

**Pandas Series**
A Pandas Series is like a column in a table. It is a 1-D array holding data of any type

Importing pandas

```python
import numpy as np
import pandas as pd
```

Series for list

```python
fruits = ['apple', 'orange', 'mango', 'kiwi']
pd.Series(fruits)
```

by defalut series take string as an object

```python
#number
numbers = [1, 2, 3, 4, 5]
pd.Series(numbers)
```

```python
# custom indexing
fruits = ['apple', 'orange', 'mango', 'kiwi']
days = ['monday', 'tuesday', 'wednesday', 'thursday']
pd.Series(fruits, days) # fruits will be value and days will become as index
```

```python
#naming a series
fruits = ['apple', 'orange', 'mango', 'kiwi']
days = ['monday', 'tuesday', 'wednesday', 'thursday']
pd.Series(fruits, days, name='fruits')
```
