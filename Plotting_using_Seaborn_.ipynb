```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
# relplot -> figure level -> square shape
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter',hue='sex',style='time')
```

```python
countriesdata = px.data.gapminder()   # build in dataset
countriesdata.head()
```

```python
pak = countriesdata[countriesdata['country'] == 'Pakistan']
pak
```

```python
#using lineplot
sns.lineplot(data=pak,x='year',y='lifeExp')
```

```python
# using relpplot
sns.relplot(data=pak, x='year', y='lifeExp', kind='line')
```

```python
# hue -> style
asiacountries = countriesdata[countriesdata['country'].isin(['India','China','Pakistan'])]
asiacountries
```

```python
#relplot
sns.relplot(kind='line', data=asiacountries, x='year', y='lifeExp', hue='country')
```

```python
#lineplot
sns.lineplot(data=asiacountries, x='year', y='lifeExp', hue='country')
```

```python
# facet plot -> figure level function -> work with relplot
# it will not work with scatterplot and lineplot
sns.relplot(data=tips, x='total_bill', y='tip', kind='line', col='sex')
```

```python
countriesdata
```

```python
#col wrape

sns.relplot(data = countriesdata, x = 'lifeExp' , y = 'gdpPercap', kind = 'scatter', col = 'year', col_wrap=3)
```

Plots under distribution plot

histplot

kdeplot

rugplot






```python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
tips
```

```python
sns.histplot(data=tips, x = 'total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
#bins level
sns.displot(data=tips, x='total_bill', kind='hist', bins=2)
```

```python
# It’s also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables

# countplot
sns.displot(data=tips, x='day', kind='hist')
```

```python
# hue parameter
sns.displot(data=tips, x='tip', kind='hist',hue='sex')
```

```python
# element -> step
sns.displot(data=tips, x='tip', kind='hist',hue='sex',element='step')
```

```python
# faceting using col and row -> not work on histplot function
sns.displot(data=tips, x='tip', kind='hist',col='sex',element='step')
```

```python
# kdeplot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# hue -> fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height=6,aspect=2)
```

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# Heatmap

# Plot rectangular data as a color-encoded matrix
temp_df = countriesdata.pivot(index='country',columns='year',values='lifeExp')

# axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# annot
temp_df = countriesdata[countriesdata['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5, cmap='summer')
```
